```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.forbes.com/billionaires/'
```

```python
r = requests.get(url)
html = r.content
```

```python
soup = BeautifulSoup(html,'html.parser')
#print(soup)
soup.prettify()
```

```python
title = soup.title
print(title.string)
```

```python
print(soup.find_all("h1",class_="title__subheading"))
#right_table
```
